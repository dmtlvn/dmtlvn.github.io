# Preamble

I was playing with [DreamBooth](https://github.com/TheLastBen/fast-stable-diffusion) one day. In case you missed it, it's a tool which fine-tunes Stable Diffusion text-to-image model on your photos, so you can then create beautiful portraits. And I believe it is truly a *"future's future future"* except one thing: training time. To create a custom model you need to wait for about 40 minutes on a Tesla T4 which is an exact opposite of fun.  

So I started thinking on how to speed up the fine-tuning, ideally without interfering with the training procedure too much and preserve image quality as much as possible. I ran a training code through a `line_profiler` to find out that it is pretty well optimized from the beginning. But not perfectly. Besides some other things a particular line caught my eye:
```
Line #  Hits        Time   Per Hit  % Time  Line Contents
==============================================================
   736   100  41914431.8  419144.3    10.1  accelerator.clip_grad_norm_(params_to_clip, args.max_grad_norm)
```
A substantial chunk of training time (10%) was spent on clipping gradient norms, and I had a great desire to remove it. So my first idea was to check if the gradient ever gets clipped, and if it doesn't then this line can be safely removed. However I have found exactly the opposite: the gradient was always getting clipped, because its raw norm was ~10<sup>3</sup> while the threshold was set to 1. Then my second though was: *"Wait a minute!"* The model is trained with the Adam optimizer, which normalizes gradients element-wise, and while it is not completely the same as dividing by the norm, it should be close enough to mask the effect of permanent gradient clipping. I ran a quick test with grad clip turned off, and it worked like a charm: training stability wasn't affected, quality of the images stayed identical, while 10% of waste time was removed - case closed, right?

# Down the rabbit hole

Unfortunately, no, because I looked into the distribution of weight updates during training. And they were indeed very similar with and without grad clip: they were converging to the same values, and the only difference was a slightly higher variance with grad clip turned off, which could come from exponential smoothing of variances in the Adam. This made me thinking: what is the difference between the distributions of weight updates using raw gradients vs. clipped gradients? I decided to check for an analytical answer to this question, and it happened to be a statistical rabbit hole which consumed the rest of my weekend. And now I'm asking myself: *Why the hell did I do that?*

Any statistic of the weight update requires a fixed model for the distribution of the gradient. And to pick such a model, I wanted to look at the distribution of gradients to answer some questions:

1. Often gradients are assumed to be normally distributed. Is it because they really are or because it just made someone's math easier? 
2. In a mini-batch setting with sensible initialization there's no reason for the gradient to have non-zero mean, because a non-zero mean would imply a preferred direction of the trajectory, which definitely is true in a full-batch regime, but in the stochastic setting it should wash out. So is this intuition really correct?
3. Parameters in the same layer have a good reason (and has been shown) to be correlated, but as the networks gets larger, are the correlations of the individual gradient components preserved and don't go away for larger networks?

# Setup

To answer these questions we are going to study several model architectures for classification to check if our findings can be generalized:
* VGG16 - a basic architecture design pattern
* ResNet18 - check if skip connections have an impact
* ViT - a trending arch

We want to make our experiments close to real life scenarios but keep minimal to save time we need a dataset, which is simple but complex enough to pose a challenge for arhcitectures above and prevent them from over-fitting right away. ImageNet is kinda big, but [Caltech256](https://data.caltech.edu/records/nyy15-4j048) with 30K images and 257 classes seems to be a reasonable choice. It has large enough images to fill receptive fields of out networks and a large variety of subjects.

We will be doing 2000 of warmup training steps first to reduce effect of initialization and grab the trajectory, then freeze the weights and sample 10 000 gradient vectors around that point. Storing 10 000 gradient vectors for a full model will clearly blow a whole through our reality, so we are going to sample only 100 000 random parameters at different convolutional layers across the network. This should cover our needs. We are also going to use Adam optimizer with default parameters throughout experiments. Batch size is set to 32 (0.01% of the data). We don't want to skew the results with large batches.

Additionally, we'll run a tiny conv net on MNIST to be able to probe the distribution along the whole learning trajectory.

````python
import numpy as np

import torch
from torch.utils.data import DataLoader, RandomSampler
from torchvision.datasets import Caltech256
from torchvision.models import resnet18, vgg16_bn
from torchvision import transforms

from grad_study.tools import collect_gradients, init_weights
import matplotlib.pyplot as plt
%matplotlib inline


def to_rgb(img):
    return img.convert('RGB')
````

````python
model = resnet18(num_classes = 257)  # uses BatchNorm by default
# model = vgg16_bn(num_classes = 257)
init_weights(model)

optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3, betas = [0.9, 0.999])

dataset = Caltech256(
    root = '../datasets/caltech256/',
    download = False,
    transform = transforms.Compose([
        transforms.Lambda(to_rgb),
        transforms.Resize(size = (128, 128)),
        transforms.RandomAffine(
            degrees = (-10, 10), 
            translate = (0.1, 0.1), 
            scale = (0.9, 1.1)
    ),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
)
sampler = RandomSampler(
    dataset, 
    replacement = False, 
    num_samples = 2**31
)
data_loader = torch.utils.data.DataLoader(
    dataset = dataset,
    sampler = sampler,
    batch_size = 32, 
    num_workers = 0,
)
gradients = collect_gradients(
    model, data_loader, optimizer, n_params = 100000, steps = 10000, warmup = 2000, cache_file = 'resnet18_wu2k_n10k_redo.pkl'
)
````

````python
import pickle

with open('vgg16_wu2k_n10k.pkl', 'rb') as file:
    gradients = pickle.load(file)
````

````python
G = np.hstack(gradients)
````

````python
import matplotlib.pyplot as plt
%matplotlib inline

### Grad mean

grad_mean = np.mean(G, axis = 0)

plt.figure(figsize = (6,3))
plt.title("Distribution of gradient means")
plt.hist(grad_mean, bins = 100, range = np.percentile(grad_mean, [5, 95]), color = 'black', alpha = 0.6, density = True);
plt.hist(grad_mean, bins = 100, range = np.percentile(grad_mean, [5, 95]), color = 'black', density = True, histtype = 'step')
plt.yticks([], []);
````

````python
from scipy.stats import ttest_1samp

pvalues = [ttest_1samp(x, 0).pvalue for x in tqdm(G.T)]
````

````python
from scipy.stats import kstest, norm

def s(x):
    return (x - x.mean()) / x.std()

pvalues = [kstest(rvs = s(x), cdf = norm.cdf).pvalue for x in tqdm(G.T)]
````

````python
plt.semilogy(np.sort(pvalues)[65000:])
````

````python
### Correlation distribution

idx = np.random.choice(np.arange(G.shape[1]), 1000, replace = False)
sub = G[:,idx]
C = np.corrcoef(sub.T)
u = np.triu(C, 1)
u = u[u != 0]

plt.figure(figsize = (6,3))
plt.title('Distribution of correlations')
plt.hist(u, bins = 100, range = np.percentile(u, [0.1, 99.9]), color = 'black', alpha = 0.6, density = True);
plt.hist(u, bins = 100, range = np.percentile(u, [0.1, 99.9]), color = 'black', density = True, histtype = 'step');
plt.yticks([], []);
````

````python
fig, axes = plt.subplot_mosaic("AABC;AADE", figsize = (16,8))
fig.suptitle('Correlation sub-matrices', fontsize = 18)

n = 20
for p in ['A', 'B', 'C', 'D', 'E']:
    idx = np.random.choice(np.arange(G.shape[1]), n, replace = False)
    sub = G[:,idx]
    C = np.corrcoef(sub.T)
    axes[p].set_title(f"Subset {p}")
    axes[p].imshow(-C, cmap = 'Greys')
    axes[p].axis('off')
````
